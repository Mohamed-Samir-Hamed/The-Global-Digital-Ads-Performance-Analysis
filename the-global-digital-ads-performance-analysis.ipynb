# The Global Digital Ads Performance

 This dataset is a 2024 global digital ads performance table with 1,800 rows (daily records) and 14 columns. Each row represents ad results for a specific date, platform (Google Ads, Meta Ads, TikTok Ads), campaign type (Search/Display/Video/Shopping), industry (5 industries), and country (7 countries). It includes core metrics like impressions, clicks, ad_spend, conversions, revenue plus efficiency metrics CTR, CPC, CPA, ROAS. The date range runs from 2024-01-01 to 2024-12-30, and the file contains no null values.

# Analysis + visuals

# What is the total performance across the whole dataset (overall KPIs)?

```python
tot_impr = df["impressions"].sum()
tot_clicks = df["clicks"].sum()
tot_spend = df["ad_spend"].sum()
tot_conv = df["conversions"].sum()
tot_rev = df["revenue"].sum()
```

```python
overall = pd.DataFrame([{
    "impressions": tot_impr,
    "clicks": tot_clicks,
    "CTR": tot_clicks / tot_impr,
    "ad_spend": tot_spend,
    "CPC": tot_spend / tot_clicks,
    "conversions": tot_conv,
    "CPA": tot_spend / tot_conv,
    "revenue": tot_rev,
    "ROAS": tot_rev / tot_spend
}])
overall
```

# Which platform performs best (ROAS, CPA, CTR, CPC) based on totals?

```python
by_platform = (df.groupby("platform", as_index=False)
                 .agg(impressions=("impressions","sum"),
                      clicks=("clicks","sum"),
                      ad_spend=("ad_spend","sum"),
                      conversions=("conversions","sum"),
                      revenue=("revenue","sum")))

```

```python
by_platform
```

```python
by_platform["CTR"]  = by_platform["clicks"] / by_platform["impressions"]
by_platform["CPC"]  = by_platform["ad_spend"] / by_platform["clicks"]
by_platform["CPA"]  = by_platform["ad_spend"] / by_platform["conversions"]
by_platform["ROAS"] = by_platform["revenue"] / by_platform["ad_spend"]

by_platform.sort_values("ROAS", ascending=False)
```

```python
plt.figure()
sns.barplot(data=by_platform, x="platform", y="CPA")
plt.title("CPA by Platform (lower is better)")
plt.tight_layout()
plt.show()
```

```python
px.bar(by_platform.sort_values("ROAS", ascending=False), x="platform", y="ROAS", title="ROAS by Platform").show()
```

# Insights

insight (1): Overall efficiency is strong: the dataset delivers ROAS  4.88 ( $4.88 revenue per $1 spend) with CPA  33.99 and CTR  3.85%.

insight (2): TikTok Ads is the top-performing platform on every key efficiency metric: highest ROAS ( 7.62), lowest CPA ( 21.67), highest CTR ( 5.53%), and lowest CPC ( 1.02).

insight (3): Google Ads drives the most spend but is the least efficient overall: lowest ROAS ( 3.47) and highest CPA ( 48.43) among platformssuggesting budget there is working harder for less return.

insight (4): Search is the best campaign type: highest ROAS ( 5.31) and lowest CPA ( 31.64) vs Display/Video/Shopping (which cluster lower on ROAS).

insight (5): India is the strongest country by efficiency: highest ROAS ( 5.50) and lowest CPA ( 30.76); USA is weakest with ROAS  4.16 and CPA  36.01.

insight (6): SaaS leads industries on ROAS ( 5.04), while Fintech lags ( 4.48)a signal of industry-level differences in conversion value or costs.

insight (7): The best ROAS combinations are dominated by TikTok across campaign types:
* Best combo: TikTok Ads + Search (ROAS  8.12)
* Worst combo: Google Ads + Display (ROAS  3.16)
* Overall: TikTok combos sit at the top, Google combos cluster at the bottom.
